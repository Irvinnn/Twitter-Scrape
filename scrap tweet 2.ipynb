```python
import GetOldTwe8ets3 as got
import pandas as pd
```

```python
df = pd.DataFrame(columns = ['Tweet'])
```

```python
df
```

```python
def get_tweets():
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('tugas online')\
        .setSince("2020-03-01")\
        .setUntil("2020-07-31")\
        .setMaxTweets(2000)
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    tweets_text = [[a.text] for a in tweets]
    return tweets_text
```

```python
text = ""
text_tweets = get_tweets()
length = len(text_tweets)

for i in range(0, length):
    text = text_tweets[i][0]
    print(text + '\n')
```

```python
for i in range(0, length):
    text = text_tweets[i][0]
    new_row = {'Tweet':text}
    df = df.append(new_row, ignore_index = True)
```
